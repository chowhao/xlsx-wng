```python
# 数据集
import pandas as pd
from matplotlib import pyplot as plt
from pylab import xticks,yticks,np
# import random
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签

test_path = '../data/cbtc/csv/new/1.hcz2cq/test.csv'
# test_path = '../data/cbtc/csv/test.csv'
test_data = pd.read_csv(test_path) / 100

test_brake = test_data.loc[:, 'brake']
test_target = test_data.loc[:, 'target']
test_speed = test_data.loc[:, 'speed']
```

```python
plt.figure(figsize=(20, 10))
# plt.title('Test Dataset')
# plt.ylabel('Train Speed m/s')
# plt.xlabel('Time / s')
plt.title('测试数据')
plt.ylabel('列车速度 m/s')
plt.xlabel('时间 / s')

data_len = len(test_data) - 1
x = 0.2 * np.linspace(0, data_len, data_len+1, endpoint=True)

# plt.plot(x, test_brake, 'b', linewidth=1, label='limit speed')
# plt.plot(x, test_target, 'r', linewidth=1, label='recommended')
# plt.plot(x, test_speed, 'y', linewidth=1, label='tracked speed')
plt.plot(x, test_brake, 'b', linewidth=1, label='制动速度')
plt.plot(x, test_target, 'r', linewidth=1, label='目标速度')
plt.plot(x, test_speed, 'y', linewidth=1, label='追踪速度')
plt.legend()
plt.savefig("test")
plt.show()
```
